# ADaM Dataset Viewer — Study ABC123

View the CDISC ADaM datasets loaded from `pharmaverseadam`.  
Datasets: **ADSL** (subjects), **ADAE** (adverse events), **ADLB** (labs)

```python
import subprocess, sys

# Install pyreadr if not available
try:
    import pyreadr
except ImportError:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "pyreadr"])
    import pyreadr

import pandas as pd
import os

data_dir = os.path.join("..", "data", "adam")

ADSL = pyreadr.read_r(os.path.join(data_dir, "ADSL.rds"))[None]
ADAE = pyreadr.read_r(os.path.join(data_dir, "ADAE.rds"))[None]
ADLB = pyreadr.read_r(os.path.join(data_dir, "ADLB.rds"))[None]

print(f"ADSL: {ADSL.shape[0]} subjects, {ADSL.shape[1]} columns")
print(f"ADAE: {ADAE.shape[0]} events,   {ADAE.shape[1]} columns")
print(f"ADLB: {ADLB.shape[0]} lab rows, {ADLB.shape[1]} columns")
```

## ADSL — Subject Level Analysis Dataset
One row per subject. Anchor dataset — all other ADaM datasets join to this via `USUBJID`.

```python
print("Columns:", list(ADSL.columns))
ADSL.head(10)
```

```python
# Treatment arm summary
ADSL.groupby("ARM").size().reset_index(name="n_subjects")
```

## ADAE — Adverse Events Analysis Dataset
One row per adverse event. Key variables: `AEDECOD` (preferred term), `AESOC` (system organ class), `TRTEMFL` (treatment-emergent flag).

```python
print("Columns:", list(ADAE.columns))
ADAE.head(10)
```

```python
# Treatment-emergent AEs only
teae = ADAE[ADAE["TRTEMFL"] == "Y"] if "TRTEMFL" in ADAE.columns else ADAE
print(f"Treatment-emergent AEs: {len(teae)}")

# Top 10 preferred terms
teae["AEDECOD"].value_counts().head(10).reset_index().rename(
    columns={"index": "Preferred Term", "AEDECOD": "Count"}
)
```

## ADLB — Laboratory Analysis Dataset
One row per lab test per visit per subject. Key variables: `PARAM`, `AVAL`, `CHG`, `ANRIND`, `ABLFL`.

```python
print("Columns:", list(ADLB.columns))
ADLB.head(100)
```

```python
# Unique lab parameters
print("Lab parameters:")
print(ADLB["PARAM"].unique() if "PARAM" in ADLB.columns else ADLB["PARAMCD"].unique())
```
